```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
df.tail()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df['Class'].value_counts() # 492 fraud transactions
```

```python
df_class = df['Class'].value_counts()
df_class
df_dic = pd.DataFrame({'index':df_class.index, 'value':df_class.values})
df_dic
colors = ['green','blue']
plt.figure(figsize=(8,8))
sbn.barplot(x=df_dic['index'], y=df_dic['value'], palette=colors)
plt.xlabel("Class")
plt.ylabel("number of transactions")
plt.show()
```

```python
data = df.loc[:,["Time","Amount"]]

# fig, axis = plt.subplot()
data.plot(kind='scatter', x="Time",y="Amount")
plt.show()
```
